```python
df.drop(columns=["Churn Category", "Churn Reason","Offer"], inplace=True)
df.sample(5)
```

```python
df.drop(columns=["Customer Status", "Churn Score", "CLTV"],inplace=True)
df.sample(3)
```

```python
contract_churn = pd.crosstab(df["Contract"], df["Churn Label"], normalize="index")

contract_churn.plot(
    kind="bar",
    title="Churn Rate by Contract Type"
)

plt.ylabel("Proportion")
plt.show()
```

```python
new_label={
    "Yes" : 1,
    "No" : 0
}

y = df["Churn Label"].map(new_label)
X = df.drop(columns=["Churn Label","Satisfaction Score","Total Charges"])
```

```python
cat_cols = X.select_dtypes(include="object").columns
print(cat_cols)
```

```python
X
```

```python
num_cols = X.select_dtypes(exclude="object").columns
print(num_cols)
```

```python
X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)
## Encoding categorical columns
```

```python
print(X_encoded.shape)
print(X_encoded.isnull().sum().sum())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_encoded[num_cols] = scaler.fit_transform(X_encoded[num_cols])
```

```python
X_encoded.describe().T[['mean', 'std']]
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(
    max_iter=1000,
    class_weight="balanced"
)

lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

y_pred = lr.predict(X_test)
y_prob = lr.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
import joblib

joblib.dump(X_test, "X_test.pkl")
joblib.dump(y_test, "y_test.pkl")
joblib.dump(lr, "logistic_churn_model.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(X_encoded.columns, "model_features.pkl")

```
